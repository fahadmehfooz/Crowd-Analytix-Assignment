# Q1.


Read the csv file (interview_scent.csv) into python program and perform the following tasks:

Find the total number of rows in the csv file, ignoring the rows where `prediction` in "None".

Find the distinct values in the `prediction` column of csv.

Find the count of each distinct value appearing in the `prediction` column of csv.

Submit the output of above question into csv file.




```python

file =  open(r'C:\Users\ABC\Downloads\interview_scent.csv', 'r')
reader = csv.reader(file)

# put the rows in csv to a list
rows = []  

# All rows appended her
for row in reader:
    rows.append(row)

# to get the count of prediction where the value is not None
rows_where_prediction_is_not_none = 0

# Dictionary that captures distinct prediction values
dictionary_ = {}

# Iterating through all the rows
for row in rows:
    # row[4] represents prediction values here and removing header row also
    if  row[4]!='predictions':
        # Counting values where None rows not present
        if  row[4]!='None':
            rows_where_prediction_is_not_none+=1
            
        if row[4] not in dictionary_:
            dictionary_[row[4]] = 1
        else:
            dictionary_[row[4]]+=1
        
# Creating a list of dictionaries to write output dictionary        
x = []
for key in dictionary_.keys():
    
    d = {}
    
    # Getting rows count where value is not none in only first row
    if len(x)==0:
        d['rows_where_prediction_is_not_none'] = rows_where_prediction_is_not_none
        
    d['distinct_words'] = key
    d['count'] = dictionary_[key]
    x.append(d)

# Writing output csv    
with open(r'C:\Users\ABC\Downloads\Outputs.csv', 'w',  newline='') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=['distinct_words', 'count', 'rows_where_prediction_is_not_none'])
    writer.writeheader()
    writer.writerows(x)
```

Take a number "N" from the user and let the user enter a set of numbers exactly equal to N, as "input_array", form a list from the given set of array elements.  With that list, write a program to print this list after removing all duplicate values with original order reserved.


Special Case 1) If any duplicates are found to raise a user-defined Exception and print "Duplicate Found, Exception Handled".

Special Case 2) Make Sure, No Matter What the input user will give make sure The Program Doesn't throw any Runtime Error.

 Constraints:
 0 < N < 10
 
Input: 

N = 5, input_array = [2,2,3,4,5]

Output: 


Duplicate Found, Exception Handled at Index 1

[5,4,3,2]



```python
class customexception(Exception):
 
    # Constructor or Initializer
    def __init__(self, value):
        self.value = value
 
    # __str__ is to print() the value
    def __str__(self):
        return(repr(self.value))
    

 
def remove_duplicate(input_array):
    
    # Creating a non duplicate list
    non_duplicate_list = list(set(input_array))
    
    # to check if length of non duplicate list is same as input list, if it is lesser than input that means we can find index of dupliacte
    if len(non_duplicate_list) < len(input_array) and len(non_duplicate_list)==1:
        try:
            # raising the exception
            pos = 1
            raise(customexception(pos))

        except customexception as error:
            print('Duplicate Found, Exception Handled at Index:  ',pos)
            print(non_duplicate_list[::-1])
    
    # When length of non duplicate list is greater than 1
    elif len(non_duplicate_list) < len(input_array) and len(non_duplicate_list)>1:
        # Iterating for the length of smaller list
        for pos in range(len(non_duplicate_list)):
            
            if non_duplicate_list[pos] == input_array[pos] and pos!= len(non_duplicate_list)-1 :
                continue
            elif non_duplicate_list[pos] == input_array[pos] and pos== len(non_duplicate_list)-1 :
                x = pos + 1
 
               
            else:
                # Breaking when on same position in both lists elements do not match
                x = pos
            try:
                # raising the exception
                raise(customexception(x))

            except customexception as error:
                print('Duplicate Found, Exception Handled at Index:  ',x)
                print(non_duplicate_list[::-1])
                break
        
                
    

    elif len(non_duplicate_list) == len(input_array):
        print(non_duplicate_list[::-1])
        
        


        
try:
    N = int(input())
    if 0<N and N<10:
        input_array = []
        for i in range(N):
            input_array.append(int(input()))
        remove_duplicate(input_array)

    else:
        raise()
except:
    print('Incorrect Input: please check the conditions ')
```
